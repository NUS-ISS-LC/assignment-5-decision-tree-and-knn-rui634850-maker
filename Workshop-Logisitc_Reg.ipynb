```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('iris-data-clean.csv')
df.head()
```

```python
print(df['class'].unique())  #check if there are other class except 'setosa', 'virginica' and 'versicolor'
```

```python
# if there are other class
#data = df[df['class'].isin(['Setosa', 'Virginica'])].copy()
df['class'] = df['class'].replace({'Setosa': 0, 'Virginica': 1, 'Versicolor': 2})
```

```python
logReg = LogisticRegression()
x = df[['sepal_length_cm', 'sepal_width_cm', 'petal_length_cm', 'petal_width_cm']]
y = df['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
logReg.fit(x_train,y_train)
```

a) Predict the outcome

```python


print(logReg.predict([[5.1, 3.5, 1.4, 0.2]]))

```

b) validate model( print accuracy_score and confusion matrix)

```python
from sklearn.metrics import accuracy_score

y_pred = logReg.predict(x_test)
acc=accuracy_score(y_test, y_pred)
print(acc)


from sklearn.metrics import confusion_matrix

print(confusion_matrix (y_test, y_pred))
#print(confusion_matrix (y_test, y_pred, labels = [1,0])) # Switching the class 0 and 1
```
